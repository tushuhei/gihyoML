# 正規分布の図示

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import pylab as plt
import matplotlib.mlab as mlab

sample = 1000
mu, sigma = 170, 5
data = np.random.normal(mu, sigma, sample)
n, bins, patches = plt.hist(data, normed=1, alpha=0.75, align='mid')
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r-', linewidth=1)
plt.xlabel('Height')
plt.ylabel('Probability')
plt.grid(True)
```

# 二項分布の図示

```python
import math
def binomial(p):
    def f(n, x):
        return comb(n, x) * (p ** x) * ((1 - p) ** (n - x))
    return f

def comb(n, r):
    return math.factorial(n) / (math.factorial(n - r) * math.factorial(r))
    
p = 0.5
L = []
trial_count = (10, 20, 30, 40, 50)
colors = ["r", "g", "b", "c", "y"]
f = binomial(p)

for n in trial_count:
    L.append([f(n, x) for x in range(n)])

for prob, color in zip(L, colors):
    plt.plot(prob, color)

plt.xlabel('Trial')
plt.ylabel('Probability')
plt.legend([str(n) for n in trial_count])
```

# ポアソン分布の図示

```python
def poisson(lambd):
    def f(x):
        return math.exp(-lambd) * float(lambd ** x) / math.factorial(x)
    return f

N = 8
L = []
score_mean = 0.8

f = poisson(score_mean)
L.append([f(k) for k in range(N + 1)])

for prob in L:
    plt.plot(range(N + 1), prob, 'r')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.grid(True)
```
