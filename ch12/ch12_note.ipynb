```python
%matplotlib inline
import numpy as np
import pylab as plt
import math
def binomial(n, p):
    def f(x):
        return comb(n, x) * (p ** x) * ((1 - p) ** (n - x))
    return f
def comb(n, x):
    return math.factorial(n) / (math.factorial(x) * math.factorial(n - x))

n = 10
p = 0.5
f = binomial(n, p)
plt.plot(range(n), [f(x) for x in range(n)])
```
